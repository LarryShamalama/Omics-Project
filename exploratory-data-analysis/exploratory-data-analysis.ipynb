```python
class Patient:
    def __init__(self, identifier, clinical_group, sample_dict):
        '''
        Can add more attributes here
        '''
        assert identifier[:6] == 'MAARS_'
        assert isinstance(sample_dict, dict)
        assert len(set(sample_dict.values()).intersection({'LES', 'NON_LES'})) > 0
        assert clinical_group in ['AD', 'PSO', 'CTRL']
        
        self.identifier     = identifier
        self.clinical_group = clinical_group
        self.sample_dict    = sample_dict
        self.summary_transcriptomes = {sample: sample_dict[sample] for sample in set(list(norm_data)).intersection(set(sample_dict.keys()))}
        
    def __str__(self):
        part1 = 'Patient_id:\t\t\t{}\nClinical Group:\t\t\t{}\nNumber of samples:\t\t{}\n'.format(self.identifier, self.clinical_group, len(self.sample_dict))
        part2 = 'Transcriptome summary:\n' + ''.join(['\t{}: {}\n'.format(_id, self.sample_dict[_id]) for _id in self.summary_transcriptomes])
        
        return part1 + part2
```

```python
# ad = 0
# not_ad = 0

list_patients = []

for patient_id in np.unique(data1['MAARS_identifier'].values):
    patient_data = data1[data1['MAARS_identifier'] == patient_id][['MAARS_identifier', 'clinical_group', 'sample_id', 'lesional']]
    assert len(set(patient_data['clinical_group'])) == 1
    
    sample_dict = {_id: _les for (_id, _les) in list(zip(patient_data['sample_id'], patient_data['lesional']))}
    
    list_patients.append(Patient(patient_data['MAARS_identifier'].values[0],
                         patient_data['clinical_group'].values[0],
                         sample_dict))
#     if set(patient_data['clinical_group']) == {'AD'}:
#         ad += 1
#     else:
#         not_ad += 1
```
